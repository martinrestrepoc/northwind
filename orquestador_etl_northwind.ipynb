```python
# Importación de librerias
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
import warnings
warnings.filterwarnings('ignore')
```

```python
df_suppliers = pd.read_sql(
    "SELECT * FROM suppliers",
    engine
)
df_suppliers
```

```python
df_orders = pd.read_sql(
    "SELECT * FROM orders",
    engine
)
df_orders
```

```python
df_order_details = pd.read_sql(
    "SELECT * FROM order_details",
    engine
)
df_order_details
```

## Transformación

- Crear dim_categorias
- Crear dim_clientes
- Crear dim_productos
- Crear dim_proveedor
- Crear dim_tiempo
- Crear fact_ordenes

```python
# Creación de la dimensión categorías
df_categories_clean = df_categories.copy()
df_categories_clean = df_categories_clean[
    ['category_id', 'category_name', 'description']
]
df_categories_clean = df_categories_clean[
    df_categories_clean['category_id'].notnull()
]
dim_categorias = df_categories_clean.copy()
dim_categorias.reset_index(drop=True, inplace=True)
dim_categorias = dim_categorias[
    ['category_id', 'category_name', 'description']
]

dim_categorias
```

```python
# Creación de la dimensión clientes

df_customers_clean = df_customers.copy()
df_customers_clean = df_customers_clean[
    ['customer_id', 'company_name', 'contact_name', 'city', 'region']
]

df_customers_clean = df_customers_clean[
    df_customers_clean['customer_id'].notnull()
]

dim_clientes = df_customers_clean.copy()
dim_clientes.reset_index(drop=True, inplace=True)
dim_clientes = dim_clientes[
    ['customer_id', 'company_name', 'contact_name', 'city', 'region']
]
dim_clientes
```

```python
# Creación de la dimensión productos
df_products_clean = df_products.copy()
df_products_clean = df_products_clean[
    ['product_id', 'product_name', 'unit_price', 'discontinued']
]

df_products_clean = df_products_clean[
    df_products_clean['product_id'].notnull()
]

df_products_clean['discontinued'] = (
    df_products_clean['discontinued'].astype(bool)
)

dim_productos = df_products_clean.copy()
dim_productos.reset_index(drop=True, inplace=True)
dim_productos = dim_productos[
    ['product_id', 'product_name', 'unit_price', 'discontinued']
]

dim_productos
```

```python
# Creación de la dimensión proveedores
df_suppliers_clean = df_suppliers.copy()
df_suppliers_clean = df_suppliers_clean[
    ['supplier_id', 'company_name', 'contact_name', 'city', 'region']
]

df_suppliers_clean = df_suppliers_clean[
    df_suppliers_clean['supplier_id'].notnull()
]

dim_proveedores = df_suppliers_clean.copy()
dim_proveedores.reset_index(drop=True, inplace=True)
dim_proveedores = dim_proveedores[
    ['supplier_id', 'company_name', 'contact_name', 'city', 'region']
]

dim_proveedores
```

```python
# Creación de la dimensión tiempo

df_time_clean = df_orders.copy()
df_time_clean = df_time_clean[['order_date']]
df_time_clean = df_time_clean[df_time_clean['order_date'].notnull()]

# Convertir a datetime
df_time_clean['order_date'] = pd.to_datetime(df_time_clean['order_date'])

df_time_clean = df_time_clean[df_time_clean['order_date'].notnull()]

# Quitar duplicados
df_time_clean = df_time_clean.drop_duplicates(subset=['order_date'])

df_time_clean['month'] = [i.month for i in df_time_clean['order_date']]
df_time_clean['day'] = [i.day for i in df_time_clean['order_date']]
df_time_clean['year'] = [i.year for i in df_time_clean['order_date']]
df_time_clean['month_name'] = [i.strftime('%B') for i in df_time_clean['order_date']]
df_time_clean['day_name'] = [i.strftime('%A') for i in df_time_clean['order_date']]
df_time_clean['quarter'] = [i.quarter for i in df_time_clean['order_date']]

dim_tiempo = df_time_clean.copy()
dim_tiempo.reset_index(drop=True, inplace=True)
dim_tiempo = dim_tiempo[
    ['order_date', 'month', 'day', 'year', 'month_name', 'day_name', 'quarter']
]

dim_tiempo
```

```python
# Creación de la tabla de hechos (ventas)
df_fact_sales = df_order_details.copy()

# 1) Limpieza mínima: quitar nulos en llaves y métricas clave
df_fact_sales = df_fact_sales.dropna(
    subset=['order_id', 'product_id', 'unit_price', 'quantity', 'discount'],
    how='any'
)

# 2) Traer info de orders (customer_id, order_date, freight)
df_fact_sales = df_fact_sales.merge(
    df_orders[['order_id', 'customer_id', 'order_date', 'freight']],
    on='order_id',
    how='left'
)

# 3) Quitar filas que no lograron hacer match con orders (por integridad)
df_fact_sales = df_fact_sales.dropna(
    subset=['customer_id', 'order_date', 'freight'],
    how='any'
)

# 4) Traer supplier_id y category_id desde products (para conectar dimensiones)
df_fact_sales = df_fact_sales.merge(
    df_products[['product_id', 'supplier_id', 'category_id']],
    on='product_id',
    how='left'
)

# 5) Quitar filas sin supplier/category (si quieres integridad estricta)
df_fact_sales = df_fact_sales.dropna(
    subset=['supplier_id', 'category_id'],
    how='any'
)

# 6) Asegurar tipo fecha para unir con dim_tiempo
df_fact_sales['order_date'] = pd.to_datetime(df_fact_sales['order_date'])

# 7) Métrica calculada: amount
df_fact_sales['amount'] = df_fact_sales['unit_price'] * df_fact_sales['quantity']

# 8) Tabla final
fact_sales = df_fact_sales.copy()
fact_sales.reset_index(drop=True, inplace=True)

fact_sales['id_sales'] = fact_sales.index + 1

fact_sales = fact_sales[
    [
        'id_sales',         # PK
        'order_id',         # número de orden (order)
        'customer_id',      # FK dim_clientes
        'product_id',       # FK dim_productos
        'supplier_id',      # FK dim_proveedores
        'category_id',      # FK dim_categorias
        'order_date',       # FK dim_tiempo
        'freight',          # medida
        'unit_price',       # medida
        'quantity',         # medida
        'discount',         # medida
        'amount'            # medida calculada
    ]
]

fact_sales
```

## Carga


```python

engine_olap = create_engine('postgresql+psycopg2://postgres:270106@localhost:5432/northwind_olap')
dim_categorias.to_sql('dim_categorias', con=engine_olap, if_exists='replace', index=False)
dim_clientes.to_sql('dim_clientes', con=engine_olap, if_exists='replace', index=False)
dim_productos.to_sql('dim_productos', con=engine_olap, if_exists='replace', index=False)
dim_proveedores.to_sql('dim_proveedores', con=engine_olap, if_exists='replace', index=False)
dim_tiempo.to_sql('dim_tiempo', con=engine_olap, if_exists='replace', index=False)
fact_sales.to_sql('fact_sales', con=engine_olap, if_exists='replace', index=False)
```
